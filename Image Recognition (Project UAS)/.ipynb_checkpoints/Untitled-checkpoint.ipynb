```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
import random
import time

from tensorflow import keras
from sklearn.metrics import jaccard_score, classification_report, f1_score
from sklearn.model_selection import train_test_split, StratifiedKFold
from matplotlib.pyplot import imread
from glob import iglob
from PIL import Image, ImageOps
from skimage import color
from skimage import io
from os import listdir
from os.path import isdir, join
from skimage.color import rgb2gray
```

```python
target_list = [name for name in listdir('FotoAI2020/') if isdir(join('FotoAI2020/', name))]
```

```python
x = []
y = []
for index, target in enumerate(target_list):
    for data in iglob('FotoAI2020/' + target + '/*.png'):
        img = Image.open(data)
        img_gray = ImageOps.grayscale(img)
        x.append(np.array(img_gray))
        y.append(index)

x = np.asarray(x)
y = np.asarray(y)
```

```python
fig, axes = plt.subplots(21, 20,figsize=(20, 20), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i, ax in enumerate(axes.flat):
    ax.imshow(x[i], cmap="gray")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=None)

print(x_train.shape)
print(x_test.shape)
print(np.unique(y_train, return_counts=True))
print(np.unique(y_test, return_counts=True))
```

```python
plt.imshow(x_train[150], cmap='gray')
```

```python
imh = 100
imw = 100
imcol_space = 1

model = keras.Sequential([
    keras.layers.Reshape(target_shape=(imh * imw * imcol_space,), input_shape=(imh, imw, imcol_space)),
    keras.layers.Dense(units=512, activation='relu'),
    keras.layers.Dense(units=256, activation='relu'),
    keras.layers.Dense(units=128, activation='relu'),
    keras.layers.Dense(units=40, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
              metrics=['acc'])

callback_acc = tf.keras.callbacks.EarlyStopping(
    monitor='acc', min_delta=0.01, patience=3, verbose=0, mode='max',
    baseline=None, restore_best_weights=True
)

start = time.time()
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=50,
    batch_size=16
)
runtime = time.time() - start
print('Runtime : ', runtime, 's')
```
